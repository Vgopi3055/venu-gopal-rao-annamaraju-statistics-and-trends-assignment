```python
# Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Reading the dataset
df = pd.read_csv("E:/walmart dataser/Walmart_sales.csv")
```

```python
# Creating a copy of dataset
df1=df.copy()
```

```python
# Describe the Dataset
df.describe()
```

```python
# Displaying the Dataset
df
```

```python
# Conversion of date column to datetime dataframe
df['Date']=pd.to_datetime(df['Date'],dayfirst=True)
```

```python
# Extracting Month and Year by using pandas DatetimeIndex()
df['month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
```

```python
# correlation of dataset
df.corr()
```

```python
# Grouping the Dataset in Store and Year by using mean function.
df1=df.groupby(['Store', 'Year']).mean()
```

```python
# Displaying the Dataset
df1
```

```python
# Plotting a histogram of weekly sales
plt.figure(figsize=(10, 6))
plt.hist(df['Weekly_Sales'], bins=20, color='purple', edgecolor='black')

# Adding a title
plt.title('Histogram of Weekly Sales')

# Adding the labels
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')

# Display
plt.show()
```

```python
# Importing a new library 
import matplotlib as mpl
```

```python
#plotting a line graph
weekly_sales_set = set(df['Year'])

# Create a figure with a specified figure size
plt.figure(figsize=(45, 17))
for year in weekly_sales_set:
     selected_data = df.loc[df['Year'] == year]
     plt.plot(selected_data['Store'], selected_data['Weekly_Sales'],marker = 'o', label=year, linestyle='-')
mpl.rcParams['font.size'] = 25

# Adding a title
plt.title('Analysis of weekly sales data over time')

# Adding labels
plt.xlabel('Store')
plt.ylabel('Weekly_Sales')

# Display
plt.legend()
plt.show()
```

```python
# Plotting  a heatmap 
plt.figure(figsize=(20,10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Adding a title
plt.title('Correlation Matrix of walmart weekly sales ')

# Display
plt.show()
```
